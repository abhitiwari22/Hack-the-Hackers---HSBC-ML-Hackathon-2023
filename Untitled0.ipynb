Feature Engineering
Generating new features from V5

```python
df[["V5",'Time']]=df["V5"].str.split(" ",n=1,expand=True)
```

```python
# Converting Total time into seconds
def get_sec(time_str):
    """Get Seconds from time."""
    h, m, s = time_str.split(':')
    return int(h) * 3600 + int(m) * 60 + int(s)
# #Applying get seconds gunction transaction time
# df["transaction_time"]=data_with_index_v1["transaction_time"].apply(lambda x:get_sec(x))
```

```python
df["Time"]=df["Time"].apply(lambda x:get_sec(x))
```

```python
df['V11'].value_counts()
```

```python
df['V5']=pd.to_datetime(df['V5'])
```

```python
df['year']=df['V5'].dt.year
df['month']=df['V5'].dt.month
df['day']=df['V5'].dt.day
```

```python
#Dropping Zero Variance features like V2
df.drop(['V2'],axis=1,inplace=True)
```

```python
df.drop(['V5','V16'],axis=1,inplace=True)
```

```python
obj_cols=df.select_dtypes(include='object').columns
```

```python
obj_cols
```

```python
# Use `LabelEncoder` to encode labels with value between 0 and n_classes-1.
#partner_category
le = LabelEncoder()

for col in obj_cols:
  df[col]=le.fit_transform(df[col])
```

```python
plt.figure(figsize=(4, 3))
sns.countplot(data=df, x="Target")
```

```python
X=df.drop(['Target'],axis=1)
y=df['Target']
```

```python
#split the data into train and test 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

Model building with decision tree

```python
clf = DecisionTreeClassifier(random_state=123) 
# Train Decision Tree Classifer
clf.fit(X_train,y_train)
DecisionTreeClassifier(random_state=123)
y_pred_train = clf.predict(X_train)
y_pred = clf.predict(X_test)
f1_score(y_test, y_pred)
```

```python
conf_matrix = confusion_matrix(y_true=y_test, y_pred=y_pred)
#
# Print the confusion matrix using Matplotlib
#
fig, ax = plt.subplots(figsize=(5, 5))
ax.matshow(conf_matrix, cmap=plt.cm.Oranges, alpha=0.3)
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        ax.text(x=j, y=i,s=conf_matrix[i, j], va='center', ha='center', size='xx-large')
 
plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show()
```
